# Hugging Face Tasks

```python
from transformers import pipeline
#-----------------------------------------------#
#               NLP TASKS                       #
#-----------------------------------------------#

'''
1. Text Classification: Assigning a category to a piece of text.
Sentiment Analysis
Topic Classification
Spam Detection
'''

classifier = pipeline("text-classification")

'''
2. Token Classification: Assigning labels to individual tokens in a sequence.
Named Entity Recognition (NER)
Part-of-Speech Tagging
'''

token_classifier = pipeline("token-classification")

'''
3. Question Answering: Extracting an answer from a given context based on the question.
'''

question_answerer = pipeline("question-answering")

'''
4. text Generation: Generating text based on a given prompt.
'''

text_generator = pipeline("text-generation")

'''
5. Summarization: Generating a concise summary of a longer text.
'''

summarizer = pipeline("summarization")

'''
6. Translation: Converting text from one language to another.
'''

translator = pipeline("translation", model = "Helsinki-NLP/opus-mt-en-fr")

'''
7. Text2Text Generation: General purpose text transformation, including summarization and translation.
'''

text2text_generator = pipeline("text2text-generation")

'''
8.Fill-Mask: Predicting the masked token in a sequence.
'''

fill_mask = pipeline("fill-mask")

'''
9. Feature Extraction: Extracting hidden states or features from text
'''

feature_extraction = pipeline("feature-extraction")

'''
10. Sentence Similarity: Measuring the similarity between two sentences.
'''

sentence_similarity = pipeline("sentence-similarity")



#-----------------------------------------------#
#          COMPUTER VISION TASKS                #
#-----------------------------------------------#

'''
1. Image Classification: Assigning a label or class to an image.
'''

image_classifier = pipeline("image-classification")

'''
2. Object Detection: Identifying and localizing objects within an image.
'''

object_detector = pipeline("object-detection")

'''
3. Image Segmentation: Assigning a mask or label to each pixel in an image.
'''

image_segmenter = pipeline("image-segmentation")

'''
4. Image Generation: Generating new images based on a given prompt.
'''

image_generator = pipeline("image-generation")

'''
5. Image Captioning: Generating a textual description of an image.
'''

image_captioner = pipeline("image-captioning")

#-----------------------------------------------#
#          SPEECH PROCESSSING TASKS             #
#-----------------------------------------------#

'''
1. Automatic Speech Recognition (ASR): Converting speech to text.
'''

speech_recognizer = pipeline("automatic-speech-recognition")

'''
2. Speech Translation: Translating speech from one language to another.
'''

speech_translator = pipeline("speech-to-speech")

'''
3. Audio Classification: Classifying audio signals into predefined categories
'''

audio_classifier = pipeline("audio-classification")

'''
4. Audio Transcription: Converting spoken language to text.
'''

audio_transcriber = pipeline("audio-transcription")


#-----------------------------------------------#
#            MULTIMODAL TASKS                   #
#-----------------------------------------------#

'''
1. Image Captioning: Generating a textual description of an image.
'''

image_captioner = pipeline("image-to-text")

'''
2. Visual Question Answering(VQA): Answering question about the content of an image
'''

#-----------------------------------------------#
#             OTHER TASKS                       #
#-----------------------------------------------#

'''
1. Table Question Answering: Answering questions based on tabular data.
'''

table_qa = pipeline("tablee-question-answering")

'''
2. Document Question Answering: Extracting answers from documents like PDFs.
'''

doc_qa = pipeline("document-question-answering")

'''
3. Time Series Forecasting: Predicting future values in time series data.
'''

time_series_forecaster = pipeline("time-series-forecasting")
```

# NLP Tasks

## Sentiment Analysis

```python
from transformers import pipeline
classifier  = pipeline("sentiment-analysis")
result = classifier("I was happy with the last match result")
print(result)
```

```python
pipeline(task = 'sentiment-analysis')("I was not sure about the GOT series")
```

```python
pipeline("sentiment-analysis")("I had a great time at the event,  \
                                met some amazing people,  \
                                and truly enjoyed the experience from start to finish. \
                              ")
```

```python
pipeline(task = "sentiment-analysis", model = "facebook/bart-large-mnli")("Everyday lots of LLMs papers are published about LLMs Evaluation. \
                                                                          Lots of them looks very promising. \
                                                                          I am not sure if we can actually evaluate LLMs. \
                                                                          There is still lots to do. \
                                                                          Don't you think?")
```

```python
pipeline(task = "sentiment-analysis", model = "ProsusAI/finbert")("Everyday lots of LLMs papers are published about LLMs Evaluation. \
                                                                          Lots of them looks very promising. \
                                                                          I am not sure if we can actually evaluate LLMs. \
                                                                          There is still lots to do. \
                                                                          Don't you think?")
```

## Batch Processing

```python
classifier = pipeline(task = "sentiment-analysis")

task_list = ["I visited the new café downtown yesterday.",
"It had a nice ambiance and was not too crowded.",
"I ordered a coffee and sat there for an hour.",
"Nothing too special, just a quiet place to work."]

classifier(task_list)
```

```python
classifier = pipeline(task = "sentiment-analysis", model = "SamLowe/roberta-base-go_emotions")

task_list = ["I visited the new café downtown yesterday.",
"It had a nice ambiance and was not too crowded.",
"I ordered a coffee and sat there for an hour and its boring.",
"Bad experience, just a quiet place to work."]

classifier(task_list)
```

## TEXT Generation

```python
# Use a pipeline as a high-level helper
from transformers import pipeline

text_generator = pipeline("text-generation", model = "distilbert/distilgpt2")
generated_text = text_generator("Today is a rainy day in London",
                                truncation = True,
                                num_return_sequences = 2)
print(f"Generated_text:\n {generated_text[0]['generated_text']}")
```
